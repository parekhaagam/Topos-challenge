```python
# import libraries
import re
import requests
from bs4 import BeautifulSoup
```

```python
main_link = "https://en.wikipedia.org/"
wikipedia_page = main_link + "wiki/List_of_United_States_cities_by_population"
page = requests.get(wikipedia_page)
soup = BeautifulSoup(page.content)
```

## Collectinig data from USA cities Wiki page and extra record from each city wiki page

storing all this record in list of dictionary

```python
table = soup.find("table", { "class" : "wikitable sortable" })
table_body = table.find('tbody')
data = []
main_dict = {}
rows = table_body.find_all('tr')
column_name = [
            'Index', 'City', 'State', '2017 estimate','2010 census', 'change', '2016 land area sq mi', '2016 land area km2',
            '2016 population density sq mi', '2016 population density km2', 'location'  
              ]
for col in column_name:
    main_dict[col] = 1
    
for index, row in enumerate(rows):
    cols = row.find_all('td')
    
    cols = [ele.text.strip() for ele in cols]
    content = [ele.replace('\xa0', ' ') for ele in cols if ele]
    if len(content) > 0:
        #data.append(content)
        
        city_dict = dict(zip( column_name, content))

        city_url = main_link + row.findAll('a')[0]['href']
        print(city_url)
        city_page = requests.get(city_url)
        city_soup = BeautifulSoup(city_page.content)
        city_table = city_soup.find("table", { "class" : "infobox" })
        city_table_body = city_table.find('tbody')
        city_rows = city_table_body.find_all('tr')
        
        for city_row in city_rows:
            city_cols_d = city_row.findAll('td')
            city_cols_h = city_row.findAll('th')
            if len(city_cols_d) == 1 and len(city_cols_h) == 1:
                row_header = city_cols_h[0].text.replace('•', '').strip()
                row_header = re.sub("[\(\[].*?[\)\]]", "", row_header)
                row_content = city_cols_d[0].text.strip()
                #print(row_header,":" ,row_content)
                if row_header.strip() != 'City':
                    city_dict[row_header.strip()] = row_content
                
                if row_header.lower().strip() in main_dict:
                    main_dict[row_header.lower().strip()] += 1
                else:
                    main_dict[row_header.lower().strip()] = 1
        data.append(city_dict)
```

```python
# converting list of dictionary into panda dataframe
import pandas as pd
df = pd.DataFrame(data)
df.head()
```

```python
# rearranging columns as by default data is placed in asecending by column name
columnsList = df.columns.tolist()
cityIndex = columnsList.index('City')
stateIndex = columnsList.index('State')
orderList = ['City', 'State'] + columnsList[:cityIndex] + columnsList[cityIndex +1 :stateIndex] + columnsList[stateIndex+1:]
df = df[orderList]
df.head()
```

```python
# just setting max column display to 500
pd.set_option('display.max_columns', 500)
```

## Performaing some basic cleaning 

```python

colList =[ '2016 land area km2','2016 land area sq mi', '2016 population density km2','2016 population density sq mi']
removeString = ['km2', 'sq mi', 'km2', 'sq mi']
for index, eachCol in enumerate(colList): 
    print(removeString[index])
    df[eachCol] = df[eachCol].apply(lambda x:x.replace(removeString[index] , "").replace('/',''))
df.head()
```

## There are multiple column having same information

eg if there are two column name airport and airports but they contain same information so combining both these columns

```python
combinedColCommonName = ['zip','airport', 'area code', 'assembly' , 'counties', 'estimate', 'fips', 'finance' , 'found', 'gnis', 'elevation', 'histo', 'incorporated' ,'airport', ]
combinedColDisplayName = ['ZIP', 'Airport', 'Area code', 'Assembly Member', 'Counties', 'Estimate', 'FIPS', 'Finance Director', 'Founded by', 'GNIS', 'Elevation', 'Historic', 'Incorporated', 'Airports']

for index in range(len(combinedColDisplayName)):
    columnsList = [each for each in df.columns.tolist() if combinedColCommonName[index] in each.lower()]
    if len(columnsList) > 0:
        df[combinedColDisplayName[index]] = df[columnsList].apply(lambda x: ','.join(x.dropna()), axis=1)
        if combinedColDisplayName[index] in columnsList:
            columnsList.remove(combinedColDisplayName[index])
        df.drop(columnsList, axis = 1, inplace = True)
```

```python
df.head()
```

## Dropping column having more than 60% empty value

```python
null_counts = df.isna().sum()
null_counts = null_counts[null_counts > 0]
null_ratios = null_counts / len(df)

# Drop columns over x% null
null_thresh = .6
null_cols = null_ratios[null_ratios > null_thresh].index
df.drop(null_cols, axis=1, inplace=True)
```

```python
# dropping index column
df.drop('Index', axis = 1, inplace = True)
df.head()
```

```python
# cleaning location data 
df['location'] =df['location'].apply(lambda x:x.split('/')[0])
```

```python
# in case of zip code it contain multiple range of zip code which were too many ,
#so found one min and max zip code from all range and replace minZipCode- maxZipCode
def findMinMax(codes):
    codes = codes.replace('x','0')
    codes = re.sub("[^0123456789\–\,]","",codes)
    codeList = re.split(',|–',codes)
    codeNumberList = [int(code.strip(),10) for code in codeList if len(code.strip()) > 0 and len(code.strip()) == 5 ]
    #print(codeNumberList)
    if len(codeNumberList) > 0:
        minZipCode = min(codeNumberList)
        maxZipCode = max(codeNumberList)
        if len(str(minZipCode)) <= 5 and len(str(maxZipCode)) <= 5:
            return str(minZipCode) + '-' + str(maxZipCode)
    else:
        return ''
df['ZIP'] = df['ZIP'].apply(lambda x: findMinMax(x))
```

```python
# Data cleaning in city
df['City'] = df['City'].apply(lambda a: a.split('[')[0])
```

```python
df.drop('Assembly Member', axis = 1, inplace = True)
df.head()
```
